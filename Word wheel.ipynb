```python
from random import choice
class Wheel:
    def __init__(self, pangram: str):
        self.optional = list(pangram)
        self.middle = choice(self.optional)
        self.optional.remove(self.middle)

    def contains(self, word: str) -> bool:
        has_middle = self.middle in word
        letters_left = self.optional[:]
        letters_left.append(self.middle)
        for l in word:
            if l not in letters_left:
                return False
            letters_left.remove(l)
        return has_middle
```

```python
def valid_word(wheel: Wheel, word: str) -> bool:
    return len(word) >= 3 and word in words and wheel.contains(word)
```
